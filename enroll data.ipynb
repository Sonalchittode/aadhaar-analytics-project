```python
def clean_state(enroll_data):
    enroll_data.loc[enroll_data['state'].str.contains('West', case=False, na=False), 'state'] = 'West Bengal'
    enroll_data.loc[enroll_data['state'].str.contains('Islands', case=False, na=False), 'state'] = 'Andaman & Nicobar Islands'
    enroll_data.loc[enroll_data['state'].str.contains('Dadra', case=False, na=False), 'state'] = 'Dadra and Nagar Haveli and Daman and Diu'
    enroll_data.loc[enroll_data['state'].str.contains('Daman', case=False, na=False), 'state'] = 'Dadra and Nagar Haveli and Daman and Diu'
    enroll_data.loc[enroll_data['state'].str.contains('Puducherry', case=False, na=False), 'state'] = 'Pondicherry'
    enroll_data.loc[enroll_data['state'].str.contains('ODISHA', case=False, na=False), 'state'] = 'Odisha'
    enroll_data.loc[enroll_data['state'].str.contains('Orissa', case=False, na=False), 'state'] = 'Odisha'
    enroll_data.loc[enroll_data['state'].str.contains('Jammu', case=False, na=False), 'state'] = 'Jammu & Kashmir'
    enroll_data.loc[enroll_data['state'].str.contains('andhra pradesh', case=False, na=False), 'state'] = 'Andhra Pradesh'

    return enroll_data
```

```python
clean_state(enroll_data)
```
