```python
def clean_state_biometric(biometric):
    biometric.loc[biometric['state'].str.contains('West', case=False, na=False), 'state'] = 'West Bengal'
    biometric.loc[biometric['state'].str.contains('Islands', case=False, na=False), 'state'] = 'Andaman & Nicobar Islands'
    biometric.loc[biometric['state'].str.contains('Dadra', case=False, na=False), 'state'] = 'Dadra and Nagar Haveli and Daman and Diu'
    biometric.loc[biometric['state'].str.contains('Daman', case=False, na=False), 'state'] = 'Dadra and Nagar Haveli and Daman and Diu'
    biometric.loc[biometric['state'].str.contains('Puducherry', case=False, na=False), 'state'] = 'Pondicherry'
    biometric.loc[biometric['state'].str.contains('ODISHA', case=False, na=False), 'state'] = 'Odisha'
    biometric.loc[biometric['state'].str.contains('Orissa', case=False, na=False), 'state'] = 'Odisha'
    biometric.loc[biometric['state'].str.contains('Jammu', case=False, na=False), 'state'] = 'Jammu & Kashmir'
    biometric.loc[biometric['state'].str.contains('andhra pradesh', case=False, na=False), 'state'] = 'Andhra Pradesh'
    biometric.loc[biometric['state'].str.contains('Chhattisghar', case=False, na=False), 'state'] = 'Chhattisgarh'

    return biometric
```

```python
clean_state_biometric(biometric)
```
