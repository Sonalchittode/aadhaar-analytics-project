```python
def clean_state(demo):
    demo.loc[demo['state'].str.contains('West', case=False, na=False), 'state'] = 'West Bengal'
    demo.loc[demo['state'].str.contains('Islands', case=False, na=False), 'state'] = 'Andaman & Nicobar Islands'
    demo.loc[demo['state'].str.contains('Dadra', case=False, na=False), 'state'] = 'Dadra and Nagar Haveli and Daman and Diu'
    demo.loc[demo['state'].str.contains('Daman', case=False, na=False), 'state'] = 'Dadra and Nagar Haveli and Daman and Diu'
    demo.loc[demo['state'].str.contains('Puducherry', case=False, na=False), 'state'] = 'Pondicherry'
    demo.loc[demo['state'].str.contains('ODISHA', case=False, na=False), 'state'] = 'Odisha'
    demo.loc[demo['state'].str.contains('Orissa', case=False, na=False), 'state'] = 'Odisha'
    demo.loc[demo['state'].str.contains('Jammu', case=False, na=False), 'state'] = 'Jammu & Kashmir'
    demo.loc[demo['state'].str.contains('andhra pradesh', case=False, na=False), 'state'] = 'Andhra Pradesh'
    demo.loc[demo['state'].str.contains('Chhatisgarh', case=False, na=False), 'state'] = 'Chhattisgarh'
    demo.loc[demo['state'].str.contains('Uttaranchal', case=False, na=False), 'state'] = 'Uttarakhand'

    return demo
```
